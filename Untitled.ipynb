```python
# mengimport library yang dibutuhkan untuk pengolahan citra histogram
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("01.jpg") # membaca gambar dengan fungsi imageio
```

```python
img_height = img.shape[0] # Mengambil nilai(piksel) tinggi gambar dari dimensi gambar.
img_width = img.shape[1] # Mengambil nilai(piksel) lebar gambar dari dimensi gambar.
img_channel = img.shape[2] # Mengambil jumlah channel gambar
```

```python
img_grayscale = np.zeros(img.shape, dtype=np.uint8)  # Membuat array kosong dengan dimensi yang sama dengan gambar

for y in range(0, img_height):  # Melakukan iterasi untuk setiap baris (y) dalam gambar.
    for x in range(0, img_width):  # Melakukan iterasi untuk setiap kolom (x) dalam gambar.
        red = img[y][x][0]  # Mendapatkan nilai komponen merah dari piksel.
        green = img[y][x][1]  # Mendapatkan nilai komponen hijau dari piksel.
        blue = img[y][x][2]  # Mendapatkan nilai komponen biru dari piksel.
        gray = (int(red) + int(green) + int(blue)) / 3  # Menghitung rata-rata dari ketiga komponen warna untuk mendapatkan nilai skala keabuan.
        img_grayscale[y][x] = (gray, gray, gray)  # Mengatur nilai piksel pada posisi (y, x) menjadi nilai skala keabuan yang sama untuk ketiga komponen warna.

plt.imshow(img_grayscale)  # Menampilkan gambar skala keabuan.
plt.title("Grayscale")  # Memberikan judul pada plot gambar.
plt.show()  # Menampilkan plot gambar.
```

```python
hg = np.zeros((256))  # Membuat array kosong dengan panjang 256 untuk menyimpan histogram.

for x in range(0, 256):  # Melakukan iterasi untuk setiap nilai intensitas dari 0 hingga 255.
    hg[x] = 0  # Menginisialisasi setiap elemen histogram dengan nilai 0.

for y in range(0, img_height):  # Melakukan iterasi untuk setiap baris (y) dalam gambar.
    for x in range(0, img_width):  # Melakukan iterasi untuk setiap kolom (x) dalam gambar.
        gray = img_grayscale[y][x][0]  # Mendapatkan nilai skala keabuan piksel pada posisi (y, x).
        hg[gray] += 1  # Menambahkan 1 pada elemen histogram dengan indeks yang sesuai dengan nilai skala keabuan.
```

```python
plt.figure(figsize=(6, 4)) # mengatur ukuran figure atau bingkai
plt.plot(hg, color="black", linewidth=2.0) # # Membuat histogram dengan menggunakan data dari array hg, mengatur warna histogram menjadi hitam, dan ketebalan grafk histogram.

plt.title("Histogram") # mengatur judul
plt.show()# menampilkan plot grafik
```
