```python
# import library yang dibutuhkan
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("01.jpg") # MEMBACA GAMBAR DENGAN FUNGSI IMAGEIO
```

```python
img_height = img.shape[0]  # Mengambil nilai tinggi gambar (jumlah baris) dari dimensi gambar.
img_width = img.shape[1]  # Mengambil nilai lebar gambar (jumlah kolom) dari dimensi gambar.
img_channel = img.shape[2]  # Mengambil nilai jumlah saluran warna gambar dari dimensi gambar.
img_type = img.dtype  # Mengambil tipe data elemen piksel dalam gambar.
```

```python
# Menyiapkan 2 piksel kosong masing2 variabel untuk rekayasa citra dengan ukuran yang sama dengan ukuran asli
img_flip_horizontal = np.zeros(img.shape, img_type)
img_flip_vertical = np.zeros(img.shape, img_type)
```

```python
for y in range(0, img_height):  # Melakukan iterasi untuk setiap baris (y) dalam gambar.
    for x in range(0, img_width):  # Melakukan iterasi untuk setiap kolom (x) dalam gambar.
        for c in range(0, img_channel):  # Melakukan iterasi untuk setiap saluran warna (c) dalam gambar.
            # Memindahkan nilai piksel dari posisi semula ke posisi horizontal terbalik.
            # Menggunakan indeks (y, img_width-1-x, c) untuk memperoleh nilai piksel di posisi horizontal terbalik.
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c]
```

```python
for y in range(0, img_height):  # Melakukan iterasi untuk setiap baris (y) dalam gambar.
    for x in range(0, img_width):  # Melakukan iterasi untuk setiap kolom (x) dalam gambar.
        for c in range(0, img_channel):  # Melakukan iterasi untuk setiap saluran warna (c) dalam gambar.
            # Memindahkan nilai piksel dari posisi semula ke posisi vertikal terbalik.
            # Menggunakan indeks (img_height-1-y, x, c) untuk memperoleh nilai piksel di posisi vertikal terbalik.
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c]
```

```python
plt.imshow(img_flip_horizontal)  # Menampilkan gambar yang sudah dibalikkan secara horizontal.
plt.title("Flip Horizontal")  # Memberikan judul pada plot gambar.
plt.show()  # Menampilkan plot gambar.

plt.imshow(img_flip_vertical)  # Menampilkan gambar yang sudah dibalikkan secara vertikal.
plt.title("Flip Vertical")  # Memberikan judul pada plot gambar.
plt.show()  # Menampilkan plot gambar.
```
